```python
folder = 'data'
os.listdir(folder)

currents = []

for antibiotic_dir in os.listdir(folder):
  print(os.path.join(folder, antibiotic_dir))
  for table_name in os.listdir(os.path.join(folder, antibiotic_dir)):
      if table_name.endswith('.csv'):
        df = pd.read_csv(os.path.join(folder, antibiotic_dir, table_name))
        currents_list = df["Current, A"].tolist()
        name = table_name.split('_')
        currents_list.insert(0, name[0])
        currents_list.insert(1, name[1])
        currents.append(pd.DataFrame(currents_list))

data = pd.concat(currents, axis=1).T
```

```python
df = pd.read_csv('/Users/viktoriasazonova/Milk/data/cefazoline/cefazoline_1e-06_1157.csv')

voltages = df["Voltage, V"].tolist()
voltages.insert(0, 'CONCENTRATION')
voltages.insert(0, 'NAME')

data.columns = voltages
data
```

```python
data.to_csv('output_file.csv', index=False)
```

```python
import matplotlib.pyplot as plt

dist = data["NAME"].value_counts()

plt.figure(figsize=(8, 5))
dist.plot(kind="bar", color="pink", edgecolor="black")
plt.xlabel("Антибиотик")
plt.ylabel("Число измерений")
plt.title("Распределение числа измерений для каждого антибиотика")
plt.xticks(rotation=45)
plt.show()
```

```python
x_values = pd.read_csv('/Users/viktoriasazonova/Milk/data/cefazoline/cefazoline_1e-06_1157.csv')['Voltage, V'].tolist()

fig, axes = plt.subplots(2, 3, figsize=(20, 12))
plt.subplots_adjust(wspace=0.3, hspace=0.4)

for idx, antibiotic in enumerate(data['NAME'].unique()):

  row = data[data['NAME'] == antibiotic].iloc[0]

  y_values = pd.to_numeric(row[2:], errors='coerce').values

  ax = axes[idx // 3, idx % 3]
  ax.plot(x_values, y_values, marker='o', linestyle='-', color='pink')
  ax.set_title(antibiotic, fontsize=10)
  ax.set_xlabel('Voltage, V', fontsize=8)
  ax.set_ylabel('Current, A', fontsize=8)
  ax.grid(True)

plt.suptitle('Вольамперограммы', fontsize=14)
plt.show()
```
